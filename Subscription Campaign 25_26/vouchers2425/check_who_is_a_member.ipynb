```python
import pandas as pd
import numpy as np
```

```python
campaign = pd.read_excel('×××¡×˜×¨ ×§××¤×™×™×Ÿ ×× ×•×™×™× 2025_26.xlsx', sheet_name= 'vouchers 2425')

seasontickets = pd.read_csv('seasontickets.csv')
seasontickets = seasontickets[seasontickets['Status'] == 'Active']
seasontickets['assign using  ID number'] = seasontickets['assign using  ID number'].astype(str)
```

```python
campaign['User Full Name'] = campaign['User first name'].str.strip() + ' ' + campaign['User last name'].str.strip()
campaign['Transaction Owner Full Name'] = campaign['Transaction owner first name'].str.strip() + ' ' + campaign['Transaction owner last name'].str.strip()

campaign = campaign.rename(columns={
    'Identifierâ€': 'User Identifierâ€',
    'Transaction owner identifier': 'Transaction Owner Identifier'
})

# make it a list of column names instead of one long string
campaign = campaign[[
    'Voucher id',
    'Voucher name',
    'Voucher number',
    'Event name',
    'Using date',
    'User Full Name',
    'User Identifierâ€',
    'User email',
    'Transaction Owner Full Name',
    'Transaction Owner Identifier',
    'Transaction owner email',
    'Discount price',
    'Seat',
    'Price',
    'Base price',
    'Value',
    'Price area',
    'Zip code',
    'Phone',
    'Delivery type',
    'Package name',
    'Voucher Price',
    'Voucher available',
    'Used transaction id',
    'Ticket voucher value',
    'Merchandise voucher value',
    'Type',
    'Event Id',
    'SeasonTicketId',
    'MerchandiseId',
    'Season ticket name',
    'MerchandiseName',
    'Expiration date',
    'Status'
]]

campaign
```

```python
# --- 0) Build helper list of final campaign columns (your order) ---
campaign_cols = [
    'Voucher id','Voucher name','Voucher number','Event name','Using date',
    'User Full Name','User Identifierâ€','User email',
    'Transaction Owner Full Name','Transaction Owner Identifier','Transaction owner email',
    'Discount price','Seat','Price','Base price','Value','Price area','Zip code','Phone',
    'Delivery type','Package name','Voucher Price','Voucher available','Used transaction id',
    'Ticket voucher value','Merchandise voucher value','Type','Event Id','SeasonTicketId',
    'MerchandiseId','Season ticket name','MerchandiseName','Expiration date','Status'
]

# --- 2) Keep only the columns you listed (in that order) ---
campaign = campaign[campaign_cols]

# --- 3) Normalize blanks (your data has "Missing value" strings) ---
for col in ['User Full Name', 'Transaction Owner Full Name']:
    campaign[col] = campaign[col].replace({'Missing value': np.nan}).astype('string').str.strip()

for col in ['User Identifierâ€', 'Transaction Owner Identifier']:
    campaign[col] = campaign[col].replace({'Missing value': np.nan})

# --- 4) Create per-person rows while preserving ALL columns ---
user_df = campaign.copy()
user_df['Person Full Name'] = user_df['User Full Name']
user_df['Person Identifier'] = user_df['User Identifierâ€']
user_df['Role'] = 'user'

owner_df = campaign.copy()
owner_df['Person Full Name'] = owner_df['Transaction Owner Full Name']
owner_df['Person Identifier'] = owner_df['Transaction Owner Identifier']
owner_df['Role'] = 'owner'

people_long = pd.concat([user_df, owner_df], ignore_index=True)

# --- 5) Drop empty person rows ---
people_long = people_long.dropna(subset=['Person Full Name', 'Person Identifier'])
people_long = people_long[people_long['Person Full Name'].str.strip() != '']

# --- 6) Remove duplicate person within the same voucher (same name+id+voucher) ---
people_long = people_long.drop_duplicates(
    subset=['Voucher id', 'Person Full Name', 'Person Identifier']
).reset_index(drop=True)

# drop old columns (safe if some are missing)
people_long = people_long.drop(
    columns=['User Full Name','User Identifierâ€','Transaction Owner Full Name','Transaction Owner Identifier', 'User email'],
    errors='ignore'
)

# move the two new columns to the desired spot (after 'Using date')
anchor = 'Using date'
pos = people_long.columns.get_loc(anchor) + 1

# insert in reverse order so final order is [Full Name, Identifier]
people_long.insert(pos, 'Person Identifier', people_long.pop('Person Identifier'))
people_long.insert(pos, 'Person Full Name', people_long.pop('Person Full Name'))

people_long['Person Identifier'] = people_long['Person Identifier'].astype(str)
people_long
```
