```python
import pandas as pd
```

```python
game_files = [
    'AuthorizedAttendanceReportCsv_Round 2_ Hapoel Tel Aviv .csv',
    'AuthorizedAttendanceReportCsv_Round 4_ Hapoel Holon .csv',
    'AuthorizedAttendanceReportCsv_Round 8_ Hapoel Afula .csv',
    'AuthorizedAttendanceReportCsv_ 专 9_ 驻注 状拽 状 专砖 -驻注  注.csv',
    'AuthorizedAttendanceReportCsv_ 专11_ 驻注 状拽 状 专砖 - 转".csv',
    'AuthorizedAttendanceReportCsv_转 专 住 专 17_ 注  .csv',
    'AuthorizedAttendanceReportCsv_转 专 住 专 18_ 专爪  .csv',
    'AuthorizedAttendanceReportCsv_转 专 住 专 20_ 驻注 驻 .csv',
    'AuthorizedAttendanceReportCsv_专 22_ 爪专 注专 转 .csv',
    'AuthorizedAttendanceReportCsv_专 24_ 驻注 专 砖注 -  .csv',
    'AuthorizedAttendanceReportCsv_专 26_ 注专 住 爪 .csv',
    'AuthorizedAttendanceReportCsv_专注 专 1_   注专 专转  .csv',
    'AuthorizedAttendanceReportCsv_专注 专 3_   注专 专转  .csv',
    'AuthorizedAttendanceReportCsv_爪 专 砖拽 2_ 驻注 转  .csv'
]

columns_needed = [
    'assign using  ID number',
    'Gender', 'Age', 'Email',
    'First name', 'Last name',
    'STRefID',
    'City',
    'Stand', 'Area',
    'Row', 'Number', 'Price area',
    'Price type', 'Price paid', 'Phone No.',
    'Discount rule'
]

dfs = []
for file in game_files:
    df = pd.read_csv(file)
    # Select only the columns you need (ignore missing columns)
    df = df[[col for col in columns_needed if col in df.columns]]
    dfs.append(df)

all_games_df = pd.concat(dfs, ignore_index=True)

# Save to one CSV file
# all_games_df.to_csv('all_games_attendance.csv', index=False)
```
